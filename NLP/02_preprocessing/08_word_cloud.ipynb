### 텍스트 전처리

```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
spam_df = pd.read_csv('./sms_spam.csv')
spam_df.head()
```

```python
corpus = spam_df['CN'][0]
corpus
```

```python
okt = Okt()
nouns = okt.nouns(corpus)
nouns
```

```python
word_counts = Counter(nouns)
word_counts
```

```python
ko_stopwords=['은','는','이','가','및','더','수']
word_counts = {word: count for word, count in word_counts.items() if word not in ko_stopwords}
word_counts
```

### WordCloud 객체 생성

```python
word_cloud = WordCloud(
    width = 800,
    height=500,
    font_path='C:\\Windows\\Fonts\\H2MJSM',
    background_color='white'
).generate_from_frequencies(word_counts)
word_cloud
```

```python
# 시각화
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

### 전체 corpus로 WordCloud 그리기!

```python
corpus = spam_df['CN']

okt = Okt()
nouns = []
for corpus_temp in corpus:
    nouns.extend(okt.nouns(corpus_temp))
nouns
```

```python
word_count = Counter(nouns)
ko_stopward=['은','는','이','이','가','몇','더','수']
word_count = {word: count for word, count in word_count.items() if word not in ko_stopwords}
```

```python
from PIL import Image
wordcloud = WordCloud(
    width=800,
    height=800,
    font_path='C:\\Windows\\Fonts\\H2MJSM',
    background_color='white',
    mask= np.array(Image.open('./apple.jpg'))
).generate_from_frequencies(word_count)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
